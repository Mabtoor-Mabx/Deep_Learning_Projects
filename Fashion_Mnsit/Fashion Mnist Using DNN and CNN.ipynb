# Fashion Mnsit Classification 

**In This Notebook, we have to predict Fashion Mnsit Classification. For This Purpose, We used Deep Neural Network and Convolutional Neural Network and compare Accuracy of DNN and CNN**

**Using Deep Neural Network, We gave got 89% Accuracy in Training data and 87% Accuracy in Test Data**

**In Convolutional Neural Network, We have Three Different Models .  In Simple CNN model, We got 90% Accuracy . In Complex CNN Model, We have got 98% Accuracy.**

### Import Libraries

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tensorflow import keras
```

### Load Data

```python
(X_train, Y_train), (X_test, Y_test) =keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
X_train[0]
```

```python
Y_train[0]
```

### Labelling The Image 

```python
class_labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
'''
0 => T-shirt/top 
1 => Trouser 
2 => Pullover 
3 => Dress 
4 => Coat 
5 => Sandal 
6 => Shirt 
7 => Sneaker 
8 => Bag 
9 => Ankle boot '''
```

```python
plt.imshow(X_train[0], cmap='Greys')
```

### Showing Images

```python
plt.figure(figsize=(15,15))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_train[i],cmap="Greys")
  plt.axis('off')
  plt.title(class_labels[Y_train[i]]+"="+str(Y_train[i]), fontsize=20)

  '''
0 => T-shirt/top 
1 => Trouser 
2 => Pullover 
3 => Dress 
4 => Coat 
5 => Sandal 
6 => Shirt 
7 => Sneaker 
8 => Bag 
9 => Ankle boot '''
```

### Feature Scaling (Normalize The Data)

```python
X_train = X_train/255
X_test= X_test/255
```

```python
X_train[0]
```

```python
X_test[0]
```

# Build Deep Neural Network

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=[28,28]),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=10, activation='softmax')
])
```

```python
model.summary()
```

### Compile The Model

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train, epochs=10)
```

**As we have seen, we are getting 89% Accuracy in Training data. Lets Evaluate The Model** 

### Evaluate The Model 

```python
model.evaluate(X_test,Y_test)
```

**WOW! We are getting 87% Accuracy in Evaluation.  Its time to predict the Visual Form of Model**

### Visualize The Result

```python
Y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(20,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_test[i],cmap="Greys")
  plt.axis('off')
  plt.title("Actual= {} \n Predicted = {}".format(class_labels[Y_test[i]], class_labels[np.argmax(Y_pred[i])]))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, [ np.argmax(i) for i in Y_pred])
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(cm,annot=True, fmt="d")
```

### Classification Report 

```python
from sklearn.metrics import classification_report
cr = classification_report(Y_test, [ np.argmax(i) for i in Y_pred], target_names = class_labels)
print(cr)
```

# Building Convolutional Neural Network Model

```python
X_train.ndim, Y_train.ndim, X_test.ndim, Y_test.ndim
```

**As we have seen, we have Dimension of `X_train` and`X_test` is 3 but we have got shape of `(28,28,3)`. So we increase ndim to 4** 

```python
X_train = np.expand_dims(X_train,-1)
X_test = np.expand_dims(X_test,-1)
```

```python
X_train.ndim, X_test.ndim
```

**We have Successfully Change The Dimensions. Now we have to Split The dataset into training and Validation data**

### Splitting The Data

```python
from sklearn.model_selection import train_test_split 
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train , test_size=0.2, random_state=42)
```

```python
X_train.shape, X_val.shape, Y_train.shape, Y_val.shape
```

## Simple CNN Model

```python
#Building CNN model
cnn_model_1 = keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
])
```

```python
cnn_model_1.summary()
```

### Compile The Model

```python
cnn_model_1.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

### Train The Model 

```python
# train cnn model
cnn_model_1.fit(X_train, Y_train, epochs=10, batch_size=512, verbose=1, validation_data=(X_val,Y_val))
```

### Evaluate The Model

```python
cnn_model_1.evaluate(X_test,Y_test)
```

### Visualize The Predictions

```python
Y_pred = cnn_model_1.predict(X_test)
```

```python
plt.figure(figsize=(20,35))

j=1
for i in np.random.randint(0, 1000,60):
  plt.subplot(10,6, j); j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap = 'Greys')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[Y_test[i]], Y_test[i], class_labels[np.argmax(Y_pred[i])],np.argmax(Y_pred[i])))
  plt.axis('off')
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16,9))
Y_pred_labels = [ np.argmax(label) for label in Y_pred ]
cm = confusion_matrix(Y_test, Y_pred_labels)

# show cm 
sns.heatmap(cm, annot=True, fmt='d',xticklabels=class_labels, yticklabels=class_labels)
```

```python
from sklearn.metrics import classification_report
cr= classification_report(Y_test, Y_pred_labels, target_names=class_labels)
print(cr)
```

## Little Complex CNN Model

```python
#Building CNN model
cnn_model_2 = keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Conv2D(filters=64, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dropout(0.25),
                         keras.layers.Dense(units=256, activation='relu'),
                         keras.layers.Dropout(0.25),
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
                         ])
```

```python
# complie the model
cnn_model_2.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
#Train the Model
cnn_model_2.fit(X_train, Y_train, epochs=20, batch_size=512, verbose=1, validation_data=(X_val, Y_val))
```

```python
Y_pred_2 = cnn_model_2.predict(X_test)
```

```python
plt.figure(figsize=(20,35))

j=1
for i in np.random.randint(0, 1000,60):
  plt.subplot(10,6, j); j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap = 'Greys')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[Y_test[i]], Y_test[i], class_labels[np.argmax(Y_pred_2[i])],np.argmax(Y_pred_2[i])))
  plt.axis('off')
```

# Very Complex CNN Model 

```python
#Building CNN model
cnn_model_3 = keras.models.Sequential([
                         keras.layers.Conv2D(filters=64, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Conv2D(filters=128, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Conv2D(filters=64, kernel_size=3, strides=(2,2), padding='same', activation='relu'),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dropout(0.25),
                         keras.layers.Dense(units=256, activation='relu'),
                         keras.layers.Dropout(0.5),
                         keras.layers.Dense(units=256, activation='relu'),
                         keras.layers.Dropout(0.25),                        
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dropout(0.10),                         
                         keras.layers.Dense(units=10, activation='softmax')
                         ])
```

```python
# complie the model
cnn_model_3.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
#Train the Model
cnn_model_3.fit(X_train, Y_train, epochs=50, batch_size=512, verbose=1, validation_data=(X_val, Y_val))
```

```python
Y_pred_3 = cnn_model_3.predict(X_test)
```

```python
plt.figure(figsize=(20,35))

j=1
for i in np.random.randint(0, 1000,60):
  plt.subplot(10,6, j); j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap = 'Greys')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[Y_test[i]], Y_test[i], class_labels[np.argmax(Y_pred_3[i])],np.argmax(Y_pred_3[i])))
  plt.axis('off')
```
